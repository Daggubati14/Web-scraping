```python
year = []
Majority = []
for i in range(len(check3)):
    year.append(check3[i].a.string.split(' ')[0])
    temp = check3[i].find_all('tr')
    Majority.append(temp[3].text.split('\n')[3])
```

```python
year_Majority = pd.DataFrame(columns = ['year_elected','Majority'])
```

```python
vinukonda.merge(year_Majority,on='year_elected',how='left')
```
