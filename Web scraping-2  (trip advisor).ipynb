```python
url = 'https://www.tripadvisor.in/Hotels-g297586-Hyderabad_Hyderabad_District_Telangana-Hotels.html'
```

```python
req = ureq(url)
article_html = req.read()
req.close()
```

```python
article_soup = soup(article_html, 'html.parser')
```

```python
print(article_soup.prettify())
```

```python
Main_div = article_soup.find_all(class_ ="prw_rup prw_meta_hsx_responsive_listing ui_section listItem")
```

```python
Main_div[0].prettify()
```

```python
hotel_name = []
min_price = []
max_price = []
Rating = []
No_of_reviews = []
perks = []
aminities = []

f = csv.writer(open('hotels.csv', 'w'))
f.writerow(['hotel_name', 'No_of_reviews','Rating','min_price','max_price','perks'])

for i in range(len(Main_div)):
    
    
    
    prices = []
    sub = Main_div[i].find_all(class_ = 'prw_rup')
    
    Rates = Main_div[i].find_all(class_ = 'prw_rup prw_common_rating_and_review_count_with_popup linespace is-shown-at-mobile')
    Rating = Rates[0].a.get('alt').split()[0]
    
    Names = Main_div[i].find_all(class_ = 'prw_rup prw_meta_hsx_listing_name listing-title')
    hotel_name = Names[0].text
    
    nreviews = Main_div[i].find_all(class_ = 'prw_rup prw_common_rating_and_review_count_with_popup linespace is-shown-at-mobile')
    No_of_reviews= nreviews[0].text.split(' ')[1]
    
    price_details = Main_div[i].find_all(class_ = 'prw_rup prw_meta_h_responsive_commerce_tvg_multi_larger_thumbnail fullHeight is-shown-at-mobile')
    price_tag = price_details[0].find_all(class_ = 'price')
    prices = []
    for k in range(len(price_tag)):
        if(len(price_tag[k].text) > 0):
            prices.append(price_tag[k].text.split()[1])   
    min_price = min(prices)
    max_price = max(prices)
    
    ab = Main_div[i].find_all(class_ = 'prw_rup prw_common_hotel_icons_list linespace is-shown-at-tablet')
    p1 = ab[0].find_all('div')
    perks = ''
    for i in range(len(p1)):
        if(len(perks) > 0):
            perks = perks + ',' +p1[i].text
        else:
            perks=p1[i].text
    
    f.writerow([hotel_name,No_of_reviews,Rating,min_price,max_price,perks])
```
